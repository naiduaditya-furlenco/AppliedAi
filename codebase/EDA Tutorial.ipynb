```python
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
iris = pd.read_csv("../data/iris.csv")
```

```python
print("Shape: {0}\n\nColumns: {1}".format(iris.shape, iris.columns))

print(iris.columns)
```

```python
print(iris["species"].value_counts())
```

```python
iris.plot(kind='scatter', x='sepal_length', y='sepal_width')
plt.show()
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(iris, hue="species", height=4) \
.map(plt.scatter, "sepal_length", "sepal_width") \
.add_legend()

plt.show()
```

```python

sns.set_style("whitegrid");
sns.pairplot(iris, hue="species", size=3)  #For analyzing higher dimension data like iris 4-D data.
plt.show()
```

```python
sns.FacetGrid(iris, hue="species", height=7).map(sns.distplot, "petal_length").add_legend()

plt.show()
```
