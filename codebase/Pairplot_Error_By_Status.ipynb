```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../data/haberman.csv")
print(data)
```

```python
#data["status"] = pd.Series(data["status"], dtype="category")
data["status"]=data["status"].astype('category')

sns.set_style("whitegrid");
sns.pairplot(data , hue="status", height=5);  #Pandas is treating status as numerical attribute giving this error.
plt.show()
```

```python
print(data["status"])
data["status"] = list(map(lambda x : "survived" if x == 1  else "died", data["status"])) 
print(data["status"])
sns.set_style("whitegrid");
sns.pairplot(data , hue="status", height=5);
plt.show()
```
