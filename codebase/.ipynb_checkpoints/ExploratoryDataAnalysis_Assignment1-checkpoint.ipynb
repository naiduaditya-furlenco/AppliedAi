```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../data/haberman.csv")

data["status"] = list(map(lambda x : "survived" if x == 1  else "died", data["status"])) 

print(data.head(10))
```

```python
'''
Describing each column in data frame.
'''

print("Columns: ", list(data.columns))

print("-----------")

print("Unique statuses: ", data["status"].unique())

print("-----------")

print(data.describe())

print("-----------")

print(data["status"].value_counts())

print("-----------")
```

#### Observations:

a). We transformed status column so that "survived" means patient lived, "died" means patient died. 

b). This mapping of status column won't have any effect on further analysis but representation of target column
    in this way is more familiar.
    
c). One problem with this dataset is that it is skewed. As can be seen from the output of value_counts for "status"   
    series, we have only 81 datapoints of non-surviving while almost 3       times (225) datapoints for surviving 
    patients. This means that whatever model we build, simple accuracy measure   
    will not be a good measure to judge our model. (For example, a model which always outputs status=1 will still   
    have given close to 75% accuracy.)

d). Due to the skewed dataset, it'll be difficult to find out the right features for prediction as we don't know the     correct measure of accuracy to be used.

#### Objective: 

Our objective is to find out patterns in data features and the inter-relationships between them which can
be useful in identifying whether a patient lived or died accurately.



```python
'''
PDF plot for patient status based on age.
'''

sns.FacetGrid(data, hue="status", height=5).map(sns.distplot, "age").add_legend();


plt.show()

    
```


#### Observations:
    i). Histograms and CDF of both surviving and non-surviving patients is extremely similar.
    ii).Hence, only age does not seem to be a good determining factor in whether a patient survives an operation 
         or doesn't.
    iii).Although it is intersting to note that patients below the age of 40 are very likely to survive the   
          treatment.



```python
'''
PDF plot for patient status based on age.
'''

sns.FacetGrid(data, hue="status", height=5).map(sns.distplot, "year").add_legend()
```

#### Observations:

    i). Just like age, year of operation gives very similar histograms and CDF for surviving and non-surviving patients.
    ii). This implies that year alone does not give a clear indication about patient status.
    iii). Similarity of plots for both types of patients over the years indicates that there likely has been 
          no significant breakthrough or development in treatment over the years in the dataset.


```python
'''
PDF plot for patient status based on nodes.
'''

sns.FacetGrid(data, hue="status", height=5).map(sns.distplot, "nodes").add_legend()

```


#### Observation:
    i). We can observe that most of the surviving patients had less number of nodes (0-5). We'll probably confirm 
        this fact further when we plot the CDF for nodes.
    ii).The PDF for surviving patients drops significantly by the time we reach 10-12 nodes. 
        Surviving patients usually don't have more than 10 nodes.
       



```python
'''
Since we determined that age and year have similar PDFs for surviving and non surviving patients, The CDF for these
two features will also be very similar. Hence, We will plot a CDF only for "nodes" feature and confirm the 
observations made with PDFs of "nodes".
'''

surviving = data[data["status"] == "survived"]

counts, bin_edges = np.histogram(surviving['nodes'], bins=25, density=True)

pdf = counts/sum(counts)

cdf = np.cumsum(pdf)

plt.plot(bin_edges[1:],cdf, label="survived");

non_surviving = data[data["status"] == "died"]

counts, bin_edges = np.histogram(non_surviving['nodes'], bins=25, density=True)

pdf = counts/sum(counts)

cdf = np.cumsum(pdf)

plt.plot(bin_edges[1:],cdf, label="died");

plt.legend()

plt.show();

```

#### Observations:

    i). More than 90% of the surviving patients had less than 10 nodes.
    ii). Close to only 65% of the non-surviving patients had <= 10 nodes.       
    iii) Interesting to note that 75% of the surviving patients had <=5 nodes. We can confirm this fact with a 
         boxplot for nodes variable and see that the 75th percentile boundary is close to 3-5 nodes and 25th 
         percentile boundary is very close to 0-2.


```python
sns.boxplot(x='status', y='nodes', data=data)
```

#### Observations:

    i) As observed with CDFs, majority of the surviving patients had less than 10 nodes and 75% of the surviving 
       patients had around 2-4 nodes.
    
    ii) Boxplot for non-surviving patients is more widespread.
    
    iii) We won't construct boxplots for age and year as we already concluded that univariate analysis on these
         features is unlikely to yield any significantly different results.


```python
sns.violinplot(x="status", y="nodes", data=data, size=5)
plt.show()

```

#### Observations:

i) Majority of the surviving patients occupy the region for lower number of nodes and the violin plot is wide for nodes 0-6.


ii) Violin plot for non-surviving patient is more widepsread and a significant thickness extends for as high as 20-30 nodes.

```python
plt.close()
sns.set_style("whitegrid");
sns.pairplot(data , hue="status", height=5);
plt.show()
```

#### Observations

1). The scatter plots for each pair of features are too overlapping  to conclude any useful feature pairing.

2). From this analysis, it seems that the features are too overlapping w.r.t the current datapoints to be used well for patient status prediction. Number of nodes alone looks like our best hope for achieving a good accuracy.

3). We might also be able to use age as a feature if the patient's age is <= 40.

4). We need different mechanism to assess our model's accuracy as the dataset is skewed.
